```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.model_selection import train_test_split
from sklearn.linear_model import PassiveAggressiveRegressor
```

```python
data = pd.read_csv("Instagramdata.csv", encoding='latin1')
print(data.head())
```

```python
import pandas as pd
import numpy as np

# Assuming 'data' is your DataFrame
# Set a seed for reproducibility
np.random.seed(42)

# Add between 1 and 2 null values in each column
for column in data.columns:
    num_nulls = np.random.randint(1, 3)  # Randomly choose between 1 and 2 null values
    null_indices = np.random.choice(data.index, num_nulls, replace=False)
    data.loc[null_indices, column] = np.nan

# Display information about null values before removal
print("Null values before removal:")
print(data.isnull().sum())

# Remove rows with null values
data = data.dropna()

# Display information about null values after removal
print("\nNull values after removal:")
print(data.isnull().sum())
```

```python
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)

model = PassiveAggressiveRegressor()
model.fit(xtrain, ytrain)
score = model.score(xtest, ytest)
print(score)

features = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
prediction = model.predict(features)
print(prediction)
```

```python
from sklearn.linear_model import LinearRegression
x = np.array(data[['Likes', 'Saves', 'Comments', 'Shares', 'Profile Visits', 'Follows']])
y = np.array(data["Impressions"])

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)


model_lr = LinearRegression()
model_lr.fit(xtrain, ytrain)
score_lr = model_lr.score(xtest, ytest)

# Print the R^2 score
print(f"Linear Regression R^2 Score: {score_lr}")
```

```python

from sklearn.linear_model import LinearRegression


model_lr = LinearRegression()
model_lr.fit(xtrain, ytrain)
score_lr = model_lr.score(xtest, ytest)
print(f"Linear Regression R^2 Score: {score_lr}")


features_lr = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
prediction_lr = model_lr.predict(features_lr)
print(f"Linear Regression Prediction: {prediction_lr}")
```

```python
from sklearn.tree import DecisionTreeRegressor

# Decision Tree Regressor
model_dt = DecisionTreeRegressor(random_state=42)
model_dt.fit(xtrain, ytrain)
score_dt = model_dt.score(xtest, ytest)
print(f"Decision Tree Regressor R^2 Score: {score_dt}")

# Predict with Decision Tree Regressor
features_dt = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
prediction_dt = model_dt.predict(features_dt)
print(f"Decision Tree Regressor Prediction: {prediction_dt}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Random Forest Regressor
model_rf = RandomForestRegressor(random_state=42)
model_rf.fit(xtrain, ytrain)
score_rf = model_rf.score(xtest, ytest)
print(f"Random Forest Regressor R^2 Score: {score_rf}")

# Predict with Random Forest Regressor
features_rf = np.array([[282.0, 233.0, 4.0, 9.0, 165.0, 54.0]])
prediction_rf = model_rf.predict(features_rf)
print(f"Random Forest Regressor Prediction: {prediction_rf}")
```
